```python
import numpy as np
import pandas as pd
import astropy.units as u
from astropy.coordinates import SkyCoord
from astroquery.gaia import Gaia
from astroquery.vizier import Vizier
import matplotlib.pyplot as plt
import random
```

```python
koi = pd.read_pickle('koi3876.pkl')
```

```python
mel2 = {}
mel2['RA'] = []
mel2['DEC'] = []
mel2['pmra'] = []
mel2['pmdec'] = []
mel2['plx'] = []
mel2['gmagFlux'] = []
mel2['gmagFluxErr'] = []
mel2['gmagNobs'] = []
mel2['expectedPoissonErr'] = []
mel2['expectedPoisNobsErr'] = []

for n in range(618, len(koi['ra'])):
    coord = SkyCoord(koi['ra'][n], koi['dec'][n], unit=(u.degree, u.degree))
    r = Gaia.query_object_async(coordinate=coord, radius=3*u.arcsec)

    try:
        loc = np.where(r['dist'] == min(r['dist']))[0][0]
    except:
        continue

    gmagFlux = r['phot_g_mean_flux'][loc]
    gmagFluxErr = r['phot_g_mean_flux_error'][loc]
    gmagNobs = r['phot_g_n_obs'][loc]
    gExpStr = np.sqrt(gmagFlux)
    gExpLoo = np.sqrt(gmagFlux) / np.sqrt(gmagNobs)

    mel2['RA'].append(koi['ra'][n])
    mel2['DEC'].append(koi['dec'][n])
    mel2['pmra'].append(koi['pmra'][n])
    mel2['pmdec'].append(koi['pmdec'][n])
    mel2['plx'].append(koi['plx'][n])
    mel2['gmagFlux'].append(r['phot_g_mean_flux'][loc])
    mel2['gmagFluxErr'].append(r['phot_g_mean_flux_error'][loc])
    mel2['gmagNobs'].append(r['phot_g_n_obs'][loc])
    mel2['expectedPoissonErr'].append(np.sqrt(r['phot_g_mean_flux'][loc]))
    mel2['expectedPoisNobsErr'].append(np.sqrt(r['phot_g_mean_flux'][loc]) / np.sqrt(r['phot_g_n_obs'][loc]))
```
