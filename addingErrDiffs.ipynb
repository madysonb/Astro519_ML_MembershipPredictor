```python
import math
```

```python
import numpy as np
import pandas as pd
import pickle
```

```python
pS = pd.read_pickle('pleiadesSet.pkl')
hS = pd.read_pickle('hyadesSet.pkl')
```

```python
# actual - expected -->
#     > 0: error too large
#    <= 0: error as expected

pS['poissonErrDiff'] = pS['gmagFluxErr'] - pS['expectedPoissonErr']
pS['poisNobsErrDiff'] = pS['gmagFluxErr'] - pS['expectedPoisNobsErr']

hS['poissonErrDiff'] = hS['gmagFluxErr'] - hS['expectedPoissonErr']
hS['poisNobsErrDiff'] = hS['gmagFluxErr'] - hS['expectedPoisNobsErr']
```

```python
pS = pS.reset_index(drop=True)
hS = hS.reset_index(drop=True)
```

```python
pm = np.asarray(pS['pmra'])
maskP = []
for p in pm:
    maskP.append(str(type(p)) == "<class 'numpy.ma.core.MaskedConstant'>")
pS = pS.drop(np.where(maskP)[0]).reset_index(drop = True)
    
pm = np.asarray(hS['pmra'])
maskH = []
for p in pm:
    maskH.append(str(type(p)) == "<class 'numpy.ma.core.MaskedConstant'>")
hS = hS.drop(np.where(maskH)[0]).reset_index(drop = True)
```

```python
gs = np.asarray(pS['gmagNobs'] == 0)
pS = pS.drop(np.where(gs == None)[0]).reset_index(drop=True)

gs = np.asarray(hS['gmagNobs'] == 0)
hS = hS.drop(np.where(gs == None)[0]).reset_index(drop=True)
```

```python
maskP = []
for p in pS['gmagFlux']:
    maskP.append(math.isnan(p))
pS = pS.drop(np.where(maskP)[0]).reset_index(drop = True)
    
maskH = []
for p in hS['gmagFlux']:
    try:
        if p == None:
            maskH.append(True)
        else:
            maskH.append(math.isnan(p))
    except:
        maskH.append(True)
hS = hS.drop(np.where(maskH)[0]).reset_index(drop = True)
```

```python
hS = hS.astype('float32')
hS['Membership'] = hS['Membership'].astype('int')

pS = pS.astype('float32')
pS['Membership'] = pS['Membership'].astype('int')
```

```python
pS.to_pickle('finalPleiadesSet.pkl')
hS.to_pickle('finalHyadesSet.pkl')
```
